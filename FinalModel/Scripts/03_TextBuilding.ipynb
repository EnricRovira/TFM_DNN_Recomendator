# Text Building

Build one text column oining description and categories

```python
import pandas as pd
import ast
import os

pd.set_option('max_colwidth', 250)
```

```python
path = os.path.join('../Data/')
print (os.listdir(path + 'FinalItems/'))
```

```python
data = pd.read_csv(path + '204kProducts.csv', sep = ';')
data.head(3)
```

## Join text

```python
df_text = pd.DataFrame(data['alternative_id'])
subset = data[['category', 'name']]
```

### Category

```python
for i, lista in enumerate(subset['category'].values):
    subset['category'].values[i] = ast.literal_eval(lista)
subset['category'] = subset['category'].apply(lambda x: ' '.join(x))
subset.head(3)
```

```python
subset.loc[43000]
```

### Brand

```python
#data['brand'] = data['brand'].str.replace(" ","_")
data.head(3)
```

### Final Column

```python
df_text['text'] = subset.apply(lambda x: ' '.join(x.dropna().astype(str)), axis = 1)
df_text.head(3)
```

```python
finaldf = pd.concat([df_text, data['brand']], axis = 1)
finaldf = finaldf[['alternative_id', 'brand', 'text']]
finaldf.head(3)
```

```python
finaldf.to_csv(path + 'Descriptions204k.csv', sep = ';', index=False)
```
